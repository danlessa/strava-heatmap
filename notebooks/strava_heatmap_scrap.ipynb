# Downloading Strava heatmap tiles

```python
import requests as req
import numpy as np
from tqdm.auto import tqdm

URL = 'https://heatmap-external-b.strava.com/tiles/all/hot/{N}/{i_x}/{i_y}@2x.png?v=19'
OUTPUT_PATH = 'output/'
FILENAME = '{N}_{i_x}_{i_y}.png'

N = 6
seq = np.arange(0, 2 ** N)
i_arr = np.transpose([np.tile(seq, len(seq)), np.repeat(seq, len(seq))])

status_code = []
for i in tqdm(i_arr):
    args = {'N': N, 'i_x': i[0], 'i_y': i[1]}
    path = OUTPUT_PATH + FILENAME.format(**args)
    url = URL.format(**args)
    r = req.get(url)
    if r.status_code == 200:
        with open(path, 'wb') as fid:
            fid.write(r.content)
    
```
